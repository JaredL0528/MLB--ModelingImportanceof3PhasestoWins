```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import pandas as pd
from pybaseball import lahman
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
```

```python
data = pd.read_csv("Data/SAL384FinalData.csv")
data['1B'] = data['H_Batting'] - data['2B'] - data['3B'] - data['HR_Batting']
data['BRAllowed'] = data['H_Pitching'] + data['BB_Pitching'] + data['HBP_Pitching']
data
```

```python
formula2 = "R_Batting ~ AB + Q('1B') + Q('2B') + Q('3B') + HR_Batting + BB_Batting + SO_Batting + GDP + HBP_Batting + SF + SB + CS"
formula1 = "W_Standings ~ R_Batting + R_Pitching"
formula3 = "R_Pitching ~ IP + BRAllowed + HR_Pitching + SO_Pitching + WP + E" 
model1 = smf.ols(formula = formula1, data = data).fit()
model2 = smf.ols(formula = formula2, data = data).fit()
model3 = smf.ols(formula = formula3, data = data).fit()
model1.summary()
model2.summary()
model3.summary()
```
